```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Read the CSV file
df = pd.read_csv("../data/Occupancy_Estimation.csv")

# Display the first few rows of the DataFrame
df

# df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
```

```python
preds = [] 
#split my datarames into two dataframes, one for training and one for test
train = df.loc[0:9000]
test = df.loc[9000:10000]

#iterate through the rows of the dataframe, if the S6_PIR field is 0 and the S7_PIR field is 0, then append 0 to the preds list, otherwise append 1
for row in df.iterrows():
    if row[1]['S6_PIR'] == 0 and row[1]['S7_PIR'] == 0:
        preds.append(0)
    else:
        preds.append(random.choice([1,2,3]))

#calculate the accuracy of the predictions
accuracy = sum(preds[i] == df.loc[i]['Room_Occupancy_Count'] for i in range(0, len(preds))) / len(preds)
accuracy
```
