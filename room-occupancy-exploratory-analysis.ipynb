```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Convert the date and time column to datetime
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Display the updated DataFrame
df.head()
```

```python
# Filter the DataFrame based on the date range
filtered_df = df[(df['Datetime'] >= '2017-12-23') & (df['Datetime'] <= '2017-12-27')]

# Create a new y-axis for Room_Occupancy_Count
ax2 = filtered_df['Room_Occupancy_Count'].plot(secondary_y=True, color='red')

# Plot the selected columns
filtered_df.plot(x='Datetime', y='S1_Sound', kind='line', ax=ax2)

# Set the x-axis label
plt.xlabel('Datetime')

# Set the y-axis labels
plt.ylabel('Sound')
ax2.set_ylabel('Room_Occupancy_Count')
plt.title('Room Occupancy Count Correlation to S1_Sound')
# Show the plot
plt.show()
```

```python
# Get the column names of the dataframe
filtered_df = df[(df['Datetime'] <= '2017-12-25')]
df = filtered_df
columns = df.columns

# Remove the 'Datetime' and 'Room_Occupancy_Count' columns from the list
columns = [col for col in columns if col not in ['Datetime', 'Room_Occupancy_Count', 'Date', 'Time']]

# Calculate the number of rows and columns for the grid
num_rows = len(columns)
num_cols = 1

# Create a new figure and subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(20, 50))

# Flatten the axs array to make it easier to iterate over
axs = axs.flatten()

# Iterate over the columns and plot each one with the Room_Occupancy_Count overlayed
for i, col in enumerate(columns):
    # Plot the column
    axs[i].plot(df['Datetime'], df[col], label=col)
    
    # Plot the Room_Occupancy_Count column overlayed
    axs[i].twinx().plot(df['Datetime'], df['Room_Occupancy_Count'], label='Room_Occupancy_Count', color='red')

    
    # Set the x-axis label
    axs[i].set_xlabel('Datetime')
    
    # Set the y-axis label
    axs[i].set_ylabel(col)
    
    # Set the title
    axs[i].set_title(f'{col} with Room_Occupancy_Count')
    
    # Add a legend
    axs[i].legend()
    
# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
